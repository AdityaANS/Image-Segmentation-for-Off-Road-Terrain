```python
#input path is the array of images that contains the paths
#location is the location that needs to be added to the base path like /content/drive.....
# output path is the new path of image with the paths added

def change_path(input_path,location):
  output_path=[]
  for image in input_path:
    x=location+image
    output_path.append(x)
  input_path=output_path
  return input_path
```
